```python
import pandas as pd
df=pd.read_csv("D:\\DATA SCIENCE\\weight-height.csv")
df
```

```python
df.describe()
```

```python
IQR=df.Height.quantile(0.75)-df.Height.quantile(0.25) # iqr= q3-q1 and q3=75%,q1=25%,q2=50%
IQR
```

```python
low_lim=df.Height.quantile(0.25) -1.5*IQR
high_lim=df.Height.quantile(0.75) -1.5*IQR

low_lim,high_lim
```

```python
df[(df.Height<high_lim) & (df.Height>low_lim)]  # to find the valid values
```

```python
df[(df.Height>high_lim) | (df.Height<low_lim)] # to find the outliers
```
