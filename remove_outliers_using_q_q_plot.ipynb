```python
#suppose we have the following datasetof 100 values:
#create dataset with 100 values that follow normal distribution.
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

data= np.random.normal(0,1,100) #here random values are created that are already normally distributed using random.normal
data[:10] #view first 10 values
```

```python
#creating a q-q plot to find the values follow normal distribution or not
fig=sm.qqplot(data, line='45') #if the points lie on 45 degree angle then they are normally distributed or not
plt.show()
```

```python
#to generate random uniform data points
data= np.random.uniform(0,1,100) #here random values are created that are uniform using random.uniform
data[:10] #view first 10 values
```

```python
# creating the graph
fig=sm.qqplot(data, line='45') #if the points lie on 45 degree angle then they are normally distributed or not
plt.show() #uniform values dont follow the 45 angle therefore not normally distributed
```
